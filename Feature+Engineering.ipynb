```python
# find columns with large ratio of missing values
check_null = cs.isnull().sum(axis=0).sort_values(ascending=False)/float(len(cs))
check_null[check_null > 0.4]
```

```python
# drop them
cs = cs.dropna(thresh=len(cs)*0.4, axis=1)
cs.shape
```

```python
# find columns with large ratio of missing values
check_null = cs.isnull().sum(axis=0).sort_values(ascending=False)/float(len(cs))
check_null[check_null > 0.4]
```

```python
cs.loc[:, cs.apply(pd.Series.nunique) == 1].columns
```

```python
# drop that column cuz it's not going to provide any useful information
cs = cs.loc[:, cs.apply(pd.Series.nunique) != 1]
cs.shape
```

```python
# missing string values
objectColumns = cs.select_dtypes(include=['object']).columns
cs[objectColumns].isnull().sum().sort_values(ascending=False)
```

```python
# drop rows with missing gender, state, age_bucket
cs = cs.loc[cs['gender'].isnull()==False, :]
cs = cs.loc[cs['state'].isnull()==False, :]
cs = cs.loc[cs['age_bucket'].isnull()==False, :]
cs[objectColumns].isnull().sum().sort_values(ascending=False)
```

```python
# missing numerical values 
cs.select_dtypes(include=[np.number]).isnull().sum().sort_values(ascending=False)
```

```python
# Mean interpolation: replace missing values using the mean along the axis.
from sklearn.preprocessing import Imputer
imr = Imputer(missing_values='NaN', strategy='mean', axis=0)

numColumns = cs.select_dtypes(include=[np.number]).columns
imr = imr.fit(cs[numColumns])   # fit data used to calculate the interpolating values
cs[numColumns] = imr.transform(cs[numColumns])  # fill in missing values with interpolating values
```
